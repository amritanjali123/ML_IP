```python
#Blend together images of differet size
img1= cv2.imread('C:/Users/HP/Downloads/Dataset/dogImages/train/071.German_shepherd_dog/German_shepherd_dog_04904.jpg')
img1= cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2= cv2.imread('C:/Users/HP/Downloads/Dataset/dogImages/train/043.Canaan_dog/Canaan_dog_03059.jpg')
img2= cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2= cv2.resize(img2, (600, 600))
```

```python
x_offset= img1.shape[1]- 600
y_offset= img1.shape[0]- 600
```

```python
rows,cols, channels= img2.shape
```

```python
roi= img1[y_offset: img1.shape[0], x_offset: img1.shape[1]]
```

```python
img2gray= cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img2gray, cmap= 'gray')
```

```python
mask_inv= cv2.bitwise_not(img2gray)
plt.imshow(mask_inv, cmap='gray')
```

```python
import numpy as np
white_background= np.full(img2.shape, 255, dtype= np.uint8)
bk= cv2.bitwise_or(white_background, white_background, mask= mask_inv)
plt.imshow(bk)
```

```python
fg= cv2.bitwise_or(img2, img2, mask= mask_inv)
plt.imshow(fg)
```

```python
final_roi= cv2.bitwise_or(roi, fg)
plt.imshow(final_roi)
```

```python
large_img= img1
small_img= final_roi
```
